```
import pandas as pd

data = pd.read_csv("/content/amazon DM.csv")
data.head()
```


Data Cleaning + Analysis Code

```
import pandas as pd
import matplotlib.pyplot as plt

# Dataset already loaded
# data = pd.read_csv("amazon.csv")

# 1. Dataset Size
print("Shape:", data.shape)

# 2. Column Names
print("Columns:", data.columns)

# 3. Missing Values Check
print("Missing Values:\n", data.isnull().sum())

# 4. Remove Missing Values
data = data.dropna()

# 5. Remove Duplicate Rows
data = data.drop_duplicates()
# Convert price columns to numeric by removing '₹' and ',' if they are still strings
if data['actual_price'].dtype == 'object':
    data['actual_price'] = data['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)
if data['discounted_price'].dtype == 'object':
    data['discounted_price'] = data['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
```

Data Cleaning

```
data.dropna(inplace=True)
data.isnull().sum()
```


Discount % calculate

```
data["Discount%"] = ((data["actual_price"] - data["discounted_price"]) / data["actual_price"]) * 100

data.head()
```

```
data[['actual_price', 'discounted_price', 'Discount%']].head()
```


Features Select

```
X = data[['actual_price', 'discounted_price', 'Discount%']]
X.head()
```


Scaling

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pd.DataFrame(X_scaled, columns=["OriginalPrice", "DiscountPrice", "Discount%"]).head()
```

Algorithm (K-Means Clustering)

```
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42)
data["Cluster"] = kmeans.fit_predict(X_scaled)

data.head()
data["Cluster"].value_counts()
```

Cluster meaning explain


Cluster 0 → Normal discount


Cluster 1 → High discount
